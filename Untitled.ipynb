## BirdsEye

```python
from skimage.transform import matrix_transform as transform

def transformImage(img, matrix):
    return cv2.warpPerspective(img, matrix, (img.shape[:2])[::-1], flags=cv2.INTER_LINEAR)


def transformPoints(pts, matrix):
    return transform(pts, matrix)


def poly2Lines(pts):
    npts = len(pts)
    return np.array([[pts[i], pts[(i+1) % npts]] for i in range(npts)]).reshape(npts*2,2)    


def undistort(img, mtx, dist, retRoi=True):
    imsize = (img.shape[:2])[::-1]
    newcameramtx, roi = cv2.getOptimalNewCameraMatrix(mtx,dist, imsize,1, imsize)
    dst = cv2.undistort(img, mtx, dist, None, newcameramtx)    
    x,y,w,h = roi
    ret = (dst, dst[y:y+h, x:x+w]) if retRoi else dst
    return ret


class BirdsEye(object):
    
    
    def CalcDstCoords(botL, topL, topR, botR):
#         d_botL = (450, botL[1])
#         d_botR = (850, botL[1])
#         d_topL = (botL[0], 50)
#         d_topR = (botR[0], 50)

        d_botL = (  botL[0], 700)
        d_botR = (  botR[0], 700)
        d_topL = (d_botL[0], 50)
        d_topR = (d_botR[0], 50)

#         d_botL = (  450, 650)
#         d_botR = (  850, 650)
#         d_topL = (d_botL[0], 50)
#         d_topR = (d_botR[0], 50)


        return d_botL, d_topL, d_topR, d_botR

    def GetTransform(botL, topL, topR, botR):
        src = np.array((botL, topL, topR, botR));
        # Get the ideal destination coords
        dst = np.array(BirdsEye.CalcDstCoords(botL, topL, topR, botR));
        # Forward transform matrix
        matrix  = cv2.getPerspectiveTransform(src.astype(np.float32), dst.astype(np.float32))
        # Inverse Transform matrix
        inv_mat = cv2.getPerspectiveTransform(dst.astype(np.float32), src.astype(np.float32))
        return src, dst, matrix, inv_mat #np.matrix(matrix), np.matrix(inv_mat)
    
    def TransformImage(image, pts=((164, 669), (607, 435), (667, 433), (1160, 657)), retMats=False, retPts=False):
        # Calculate Transformation matrices
        src, dst, mat, inv_mat = BirdsEye.GetTransform(*tuple(pts))
        # Transform Image
        ret = transformImage(image, mat)
        if retMats: ret = (ret, mat, inv_mat)
        if retPts : ret = ret + (src, dst)
        return ret
```

```python
def clrCvt(im, flag, ch):
    ch = min(2, ch)
    im2 = cv2.cvtColor(im, flag)
    return im2 if (ch < 0) else im2[:,:,ch]
    
def hls(im, ch=2):
    return clrCvt(im, cv2.COLOR_RGB2HLS, ch)

def hsv(im, ch=1):
    return clrCvt(im, cv2.COLOR_RGB2HSV, ch)
```

## Camera Calibration 

```python
class CamCalib:

    def calcCalibParams(objpoints, imgpoints, imgs):
        imsize = (imgs[0].shape[:2])[::-1]
        ret = cv2.calibrateCamera(objpoints, imgpoints, imsize, None, None)
        ret, mtx, dist, rvecs, tvecs = ret
        return mtx, dist

    def FindCorners(calib_images, nSqr=(9,6)):
        # termination criteria
        criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)

        nx, ny = nSqr
        # prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(6,5,0)
        objp = np.zeros((nx * ny,3), np.float32)
        objp[:,:2] = np.mgrid[0:nx, 0:ny].T.reshape(-1,2)

        # Arrays to store object points and image points from all the images.
        objpoints = [] # 3d points in real world space
        imgpoints = [] # 2d points in image plane.
        inds      = []
        for ind, img in enumerate(calib_images):
            gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
            # Find the chess board corners
            ret, corners = cv2.findChessboardCorners(gray, nSqr, None)
            # If found, add object points, image points (after refining them)
            if ret == True:
                # Refine the pixel points
                corners2=cv2.cornerSubPix(gray, corners, (11,11), (-1,-1), criteria)
                # update data
                objpoints.append(objp)
                imgpoints.append(corners2)    
                inds.append(ind)
        return objpoints, imgpoints, np.take(calib_images, inds, axis=0)

    def Run(calib_images, draw=False):
        # Find Corners
        objpoints, imgpoints, imgs = CamCalib.FindCorners(calib_images)    
        # Calculate Calib Matrix
        matrix, dist = CamCalib.calcCalibParams(objpoints, imgpoints, imgs)
        # Undistort Images
        undist_imgs = np.array([undistort(img, matrix, dist) for img in imgs])
        undist_imgs, undist_roi_imgs = undist_imgs[:,0], undist_imgs[:,1]

        if (draw):
            # Plot some images        
            n_plot = 5
            grid_imgs = [np.hstack((imgs[i], undist_imgs[i])) for i in sample(range(len(imgs)), n_plot)]
            plot_grid(5,1, lambda r, c, i, fig, axs: axs[r,c].imshow(grid_imgs[i]))

        return matrix, dist, objpoints, imgpoints, imgs, undist_imgs, undist_roi_imgs    

ret = CamCalib.Run(calib_images)
calib_matrix, calib_dist = ret[0], ret[1]
```

```python
class Sobel:
    def Dir_thresh(gray, sobel_kernel=9, thresh=(-180, 180)):
        # Convert to grayscale
        if len(gray.shape) == 3: gray = hls(gray)
        # Convert to radians if needed
        thresh = np.deg2rad(thresh)

        # Calculate the x and y gradients
        sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
        sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
        # Take the absolute value of the gradient direction, 
        # apply a threshold, and create a binary image result
        absgraddir = np.arctan2(np.absolute(sobely), np.absolute(sobelx))
#         binary_output =  np.zeros_like(absgraddir).astype(bool)
#         binary_output[(absgraddir >= thresh[0]) & (absgraddir <= thresh[1])] = 1

        # Return the binary image
        return absgraddir
#         return binary_output
    
    # Define a function to return the magnitude of the gradient
    # for a given sobel kernel size and threshold values
    def Mag_thresh(gray, sobel_kernel=9, mag_thresh=(30, 100)):
        # Convert to grayscale
        if len(gray.shape) == 3: gray = hls(gray)
        # Take both Sobel x and y gradients
        sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
        sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
        # Calculate the gradient magnitude
        gradmag = np.sqrt(sobelx**2 + sobely**2)
        # Rescale to 8 bit
        scale_factor = np.max(gradmag)/255 
        gradmag = (gradmag/scale_factor).astype(np.uint8) 
        # Create a binary magnitude and direction thresholds
        mask_mag = (gradmag >= mag_thresh[0]) & (gradmag <= mag_thresh[1])
        # Create a binary image
        binary_output = np.zeros_like(gradmag)
        binary_output[mask_mag] = 1
        # Return the binary image
        return binary_output
    
    def MagRaw(gray, sobel_kernel=9, mag_thresh=(30, 100)):
        # Convert to grayscale
        if len(gray.shape) == 3: gray = hls(gray)
        # Take both Sobel x and y gradients
        sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
        sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
        # Calculate the gradient magnitude
        gradmag = np.sqrt(sobelx**2 + sobely**2)
        # Rescale to 8 bit
#         scale_factor = np.max(gradmag) 
#         gradmag = (gradmag/scale_factor).astype(np.uint8) 
#         # Create a binary magnitude and direction thresholds
#         mask_mag = (gradmag >= mag_thresh[0]) & (gradmag <= mag_thresh[1])
#         # Create a binary image
#         binary_output = np.zeros_like(gradmag)
#         binary_output[mask_mag] = 1
#         # Return the binary image
        return gradmag/np.max(gradmag)
```

## Udacity Code

```python
class uda:

    def find_centers(warped, window_width=80, margin=100, ):
        window = np.ones(window_width) # Create our window template that we will use for convolutions
        # Sum quarter bottom of image to get slice, could use a different ratio
        l_sum = np.sum(warped[int(1*warped.shape[0]/4):int(3*warped.shape[0]/4),:int(warped.shape[1]/2)], axis=0)
        l_center = np.argmax(np.convolve(window,l_sum))-window_width/2
        r_sum = np.sum(warped[int(1*warped.shape[0]/4):int(3*warped.shape[0]/4),int(warped.shape[1]/2):], axis=0)
        r_center = np.argmax(np.convolve(window,r_sum))-window_width/2+int(warped.shape[1]/2)

#         plt.subplot(121), plt.plot(l_sum), plt.plot(l_center, 'rx')
#         plt.subplot(122), plt.plot(r_sum), plt.plot(r_center, 'rx')
#         plt.show()
        return l_center, r_center
        
    def find_window(warped, window_width, window_height, margin):

        def window_mask(width, height, img_ref, center,level):
            output = np.zeros_like(img_ref)
            output[int(img_ref.shape[0]-(level+1)*height):int(img_ref.shape[0]-level*height),max(0,int(center-width/2)):min(int(center+width/2),img_ref.shape[1])] = 1
            return output

        def find_window_centroids(warped, window_width, window_height, margin):

            window_centroids = [] # Store the (left,right) window centroid positions per level
            window = np.ones(window_width) # Create our window template that we will use for convolutions

            # First find the two starting positions for the left and right lane by using np.sum to get the vertical image slice
            # and then np.convolve the vertical image slice with the window template 

            # Sum quarter bottom of image to get slice, could use a different ratio
            l_sum = np.sum(warped[int(3*warped.shape[0]/4):,:int(warped.shape[1]/2)], axis=0)
            l_center = np.argmax(np.convolve(window,l_sum))-window_width/2
            r_sum = np.sum(warped[int(3*warped.shape[0]/4):,int(warped.shape[1]/2):], axis=0)
            r_center = np.argmax(np.convolve(window,r_sum))-window_width/2+int(warped.shape[1]/2)

            # Add what we found for the first layer
            window_centroids.append((l_center,r_center))

            # Go through each layer looking for max pixel locations
            for level in range(1,(int)(warped.shape[0]/window_height)):
                # convolve the window into the vertical slice of the image
                image_layer = np.sum(warped[int(warped.shape[0]-(level+1)*window_height):int(warped.shape[0]-level*window_height),:], axis=0)
                conv_signal = np.convolve(window, image_layer)
                # Find the best left centroid by using past left center as a reference
                # Use window_width/2 as offset because convolution signal reference is at right side of window, not center of window
                offset = window_width/2
                l_min_index = int(max(l_center+offset-margin,0))
                l_max_index = int(min(l_center+offset+margin,warped.shape[1]))
                l_center = np.argmax(conv_signal[l_min_index:l_max_index])+l_min_index-offset
                # Find the best right centroid by using past right center as a reference
                r_min_index = int(max(r_center+offset-margin,0))
                r_max_index = int(min(r_center+offset+margin,warped.shape[1]))
                r_center = np.argmax(conv_signal[r_min_index:r_max_index])+r_min_index-offset
                # Add what we found for that layer
                window_centroids.append((l_center,r_center))

            return window_centroids

        window_centroids = find_window_centroids(warped, window_width, window_height, margin)

        # If we found any window centers
        if len(window_centroids) > 0:

            # Points used to draw all the left and right windows
            l_points = np.zeros_like(warped)
            r_points = np.zeros_like(warped)

            # Go through each level and draw the windows 	
            for level in range(0,len(window_centroids)):
                # Window_mask is a function to draw window areas
                l_mask = window_mask(window_width,window_height,warped,window_centroids[level][0],level)
                r_mask = window_mask(window_width,window_height,warped,window_centroids[level][1],level)
                # Add graphic points from window mask here to total pixels found 
                l_points[(l_points == 255) | ((l_mask == 1) ) ] = 255
                r_points[(r_points == 255) | ((r_mask == 1) ) ] = 255

            # Draw the results
            template = np.array(r_points+l_points,np.uint8) # add both left and right window pixels together
            zero_channel = np.zeros_like(template) # create a zero color channel
            template = np.array(cv2.merge((zero_channel,template,zero_channel)),np.uint8) # make window pixels green
            warpage = np.array(cv2.merge((warped,warped,warped)),np.uint8) # making the original road pixels 3 color channels
            output = cv2.addWeighted(warpage, 1, template, 0.5, 0.0) # overlay the orignal road image with window results

        # If no window centers found, just display orginal road image
        else:
            output = np.array(cv2.merge((warped,warped,warped)),np.uint8)
        return output
```

## Image / Rect Utilities

```python
from scipy.misc import imresize
from scipy.signal import convolve2d, convolve, gaussian

def widen(roi, left=0, right=0):
    pts = np.array(roi)
    pts[:2, 0] -= left
    pts[2:, 0] += right
    return pts

def shorten(roi, top=0, bot=0):
    pts = np.array(roi)
    pts[0, 1] += bot
    pts[3, 1] += bot
    pts[1, 1] += top
    pts[2, 1] += top    
    return pts

def reshape_hwc(img):
    return img if (len(img.shape)==3) else img.reshape(img.shape + (1,))

def gray3(gray): return np.stack((gray,)*3, axis=2)

def getColor(img):
    return img if (len(img.shape)==3) else gray3(img)

def polylines(im, pts, color=[0,200,0], thickness=5, isClosed=True, copy=True):
    im = getColor(np.copy(im) if copy else im)
    return cv2.polylines(im, pts=np.array([pts], dtype=np.int32), isClosed=isClosed, color=color, thickness=thickness)
    
def polyfills(im, pts, color=[0,200,0], copy=True):
    im = getColor(np.copy(im) if copy else im)
    return cv2.fillPoly(im, pts=np.array([pts], dtype=np.int32), color=color)

def circle(im, pts, radius=10, color=[0,200,0], thickness=-1, copy=True):
    im = getColor(np.copy(im) if copy else im)
    for pt in pts: cv2.circle(im, tuple(pt), radius=radius, color=color, thickness=thickness)
    return im
    
def rect(im, topLeft, sizeWH, color=[0,200,0], thickness=5, fill=False, copy=True):
    x, y = topLeft
    w, h = sizeWH
    pts  = np.array([(x, y), (x+w,y), (x+w, y+h), (x,y+h)], dtype=np.int)
    poly = polyfills if fill else polylines
    return poly(im, pts, color=color, thickness=thickness, copy=copy, isClosed=True)
        
def apply_mask(img, mask, false_value=0):
    dst = np.copy(reshape_hwc(img))
    if not (type(false_value)==list or type(false_value)==np.ndarray): 
        false_value = [false_value]*dst.shape[2]
    for i in range(dst.shape[2]):
        dst[mask == False, i] = false_value[i]            
    return dst.reshape(img.shape)

def fig2Img ( fig, out_res=None):
    """
    @brief Convert a Matplotlib figure to a 4D numpy array with RGBA channels and return it
    @param fig a matplotlib figure
    @return a numpy 3D array of RGBA values
    """
    # draw the renderer
    fig.canvas.draw ( )
 
    # Get the RGBA buffer from the figure
    w,h = fig.canvas.get_width_height()
    buf = numpy.fromstring ( fig.canvas.tostring_argb(), dtype=numpy.uint8 )
    buf.shape = ( w, h, 4 )
     
    if (out_res is not None): buf = imresize(buf, out_res)        
    return buf[:,:,1:]
```

```python
epsilon = 10.**-10
def normalize(vec): return vec/(np.linalg.norm(vec,1) + epsilon)
def weighted_avg_filt(vec): return np.array(vec)/(np.sum(vec) + epsilon)

class History():
    
    def __init__(self, n=3):
        self.n = n
        self.data = {}
        self.sample_weights = weighted_avg_filt(gaussian(2*n+1, n/5.0, False)[n+1:])

    def get_err(self, key, sample, err_weights=None):
        if err_weights is None: err_weights = np.ones_like(sample)
        avg = self.get_avg(key)
        return np.linalg.norm((avg*err_weights) - sample, 1) if avg is not None else None
        
    def get_avg(self, key):
        if self.get(key, nsamples=self.n) is None: return None
        s = np.array(self.get(key, nsamples=self.n))
        return np.mean(s, axis=0)
#         w = self.sample_weights
#         if (len(s.shape) != len(w.shape)): w = w.reshape([len(w)] + [1 for i in s.shape[1:]])
#         return np.sum(s * w, axis=0)
        
    def add(self, key, sample, condition=True):
        if not condition: return
        if not key in self.data: 
            self.data[key] = [sample]*self.n
        else:
            self.data[key].insert(0, sample)
        
    def get(self, key, nsamples=None):
        return self.data[key][:nsamples] if key in self.data else None
        
    def add_coeffs(self, coeffs, condition=True): 
        self.add('coeffs', coeffs, condition=condition)
        
    def get_coeffs(self):
        return self.get('coeffs', nsamples=self.n)
    
    def get_avg_coeffs(self):
        return self.get_avg('coeffs')
    
    def get_err_coeffs(self, sample):
        if (len(sample) != 3): return None        
        avg = self.get_avg('coeffs')
        if avg is None: return None
        return self.calc_diff_coeff(avg, sample)
    
    def calc_diff_coeff(self, a, b):
        normfactor = np.array([0.000664, 0.541802, 426.808382])
        n_diff = (np.absolute(a - b)/normfactor).clip(0,1)
        return (np.mean(n_diff), n_diff)
```

```python
def renderAllImages(frame, ud_frame=None, be_img=None, be_sbl=None, src=None, dst=None, be_mask_pts=None):
    def _imshow(im, name="Image", cmap=None):
        plt.title(name), plt.xticks([]), plt.yticks([])
        plt.imshow(im, cmap=cmap), plt.show()
    
    ## Show Original
    _imshow(frame, "Original")

    ## Undistort
    if ud_frame is not None: _imshow(ud_frame, "Undistorted")

    ## Apply Bird's Eye Transformation
    if (be_img is not None) and (src is not None) and (dst is not None):            
        ud_poly, be_img_poly = polylines(ud_frame, src), polylines(be_img, dst)
        _imshow(np.hstack((ud_poly, be_img_poly)), "Bird's Eye Transform")
                    
        ## After Sobel-Magnitude-Thresholding
        if be_sbl is not None:
            _imshow(be_sbl, "Sobel-Mask", cmap='gray')
            _imshow(apply_mask(be_img, be_sbl), "Sobel-Mask Applied to Bird's Eye View")
            

def doAll(frame, frame_num=-1, sbl_param = (7, (25, 150)), be_pts=[], l_hist=None, r_hist=None, show_plots=False, verbose=False, draw_hist=False):    
        
    l_txtMsg, r_txtMsg = [], []
    ## Colors 
    red, green, blue = [255, 0, 0], [ 0, 255, 0], [ 0, 0, 255]
    white,  green_l, blue_l = [255,255,255], [ 51, 255,51], [ 51,255,255]
    yellow, orange, magenta = [255, 255, 0], [255, 128, 0], [255, 0, 128]

    def addMsg(msg, side, clr=green):
        '''Appends a line to the message array. 
           Adds [[message, color]]'''
        if msg is not None and len(msg) > 0:
            txt = None
            if side == 'left' : txt = l_txtMsg
            if side == 'right': txt = r_txtMsg
            if txt is not None: txt += [[msg, clr]]


    def create_curve(poly_coeff, h, ploty=None, skip=10):
        ''' Given a quadratic polynomial, calculate the curve points
            from [0, h]. Skips 10 pixels by default
            
            Returns : [N x 2] array of points (x,y).
        '''
        if len(poly_coeff) == 0: return []
        if ploty is None: ploty = np.arange(0, h).reshape(h,1)[::skip]
        curve = poly_coeff[0]*ploty**2 + poly_coeff[1]*ploty + poly_coeff[2]
        return [[int(c), int(p)] for c,p in zip(curve, ploty)] #np.hstack((curve, ploty)).astype(int)

    
    def calc_fit_confidence(lane_fit, pts, thresh=5): # PARAM_TUNING
        max_pts = 10. #PARAM_TUNING
        def num_pts_score(pts, coeffs):
            n = float(len(pts))    # Num Points 
            return min(n/max_pts, 1.) # PARAM_TUNING : score of num points found [0,1] 

        def num_inliers_score(pts, coeffs):
            n = float(len(pts))    # Num Points
            a = np.array(coeffs[0]*pts[:,1]**2 + coeffs[1]*pts[:,1] + coeffs[2])
            inliers = np.absolute(a - pts[:,0]) <= thresh
            n_inliers = float(len(np.nonzero(inliers)[0]))
            w = np.array([1.0, 0.0]) #PARAM_TUNING 
            return np.array([np.mean([float(n_inliers / n), float(n_inliers / max_pts)]), 0.])
#             return np.array([float(n_inliers / n), float(n_inliers / max_pts)]) * w

        if (len(lane_fit) != 3 or len(pts) < 3): return 0., np.array([0.,0.,0.])
        pts = np.array(pts)
        conf = np.append(num_pts_score(pts, lane_fit), num_inliers_score(pts, lane_fit))
        return conf.sum()/2., conf

    
    def fit_lane_to_curves(pts):
        ''' Given lane points detected estimates a quadratic polynomial
        
            Returns: 
                [] if no lanes were detected 
                    else 
                [x**2, x, c] params for the quadratic fit.
        '''
        # Check validity
        if (len(pts) < 2): return []
        # Convert to numpy array
        pts = np.array(pts,dtype=np.int)
        # Fit pts to 2nd order poly
        lane_fit = np.polyfit(pts[:,1], pts[:,0], 2)
        # Check error
        return [] if (len(lane_fit) != 3) else list(lane_fit)

    
    def find_peaks(patch, center, offset=0, width=None, axs=None):
        ''' Given a patch of birds eye image that is gradient thresholded
            this routine will try to find the lane centers in the image. 
            
            patch:  The image patch is expected to be either left or right halfs of
                    the original patch. 
            
            center: Center of the patch, should the center of the last patch or mid of patch.
            width:  Provide the window width of the horizontal lane search
                    Default = len(patch) * 0.4
            axs:    If show_plots is set, then pass an axs for rendering intermediate states            
        
        
            Return: array of peak points or [], [window start index, window end index] 
        '''
        
        # Determine patch length and search window width (if none)
        n = len(patch)
        if width is None: width = int(n * 0.4) # PARAM_TUNING
        # Determine window elements of the patch and calculate signal statistics 
        n_min, n_max = max(center - int(width/2), 0), min(center + int(width/2), n)
        window = patch[n_min:n_max]
        pmean, pstd = window.mean(), window.std()
        # Threshold the signal to be greater than 1stdev. 
        p = np.nonzero(window >= (pmean + pstd))[0]
        pts = []
        # Fail if no signal above threshold was found of signal stregthen is less than 1.0
        if (len(p) == 0):
            if verbose: print("len(p) == 0")
        elif (pmean < 1.0): # PARAM_TUNING
            if verbose: print("pmean < 1.0")
        else:
            # Successfully found peaks
            x = int(np.mean(p))
            y = window[x]
            pts = [x + offset + n_min, y] if (y > 0) else []
            if verbose: print("x, y = {:3.1f}, {:3.1f}  |  Pts = {:3.1f}, {:3.1f}".format(x, y, pts[0], pts[1]))
                
        if show_plots:
            print("[min, center ,max] = [{}, {}, {}],  mean|std = {:2.5f} | {:2.5f}, Pmean|Pstd = {:2.5f} | {:2.5f}".format(n_min, center, n_max, pmean, pstd, patch.mean(), patch.std()))
            axs.plot(patch)
            axs.axvspan(n_min, n_max, color='g', fill=False)
            if (len(pts)>0): axs.plot(x + n_min, y, 'rx')
        return pts, [int(n_min + offset), int(n_max+offset)]
    
    
    def find_left_right_peaks(patch, centers):
        ''' Given a patch of BirdsEye sobel thresholded image, and 
            best estimate of the center of the road lane, finds the 
            lane peaks for left and right lane-marker peaks.
            
            Returns: left, right, n, l_winx, r_winx
                left:   left peak points,   
                right:  right peak point, 
                n:      calculated center of the patch, that splits left/right lanes
                l_winx: left_window_index_range, 
                r_winx: right_window_index_range
        '''
        patch = np.copy(patch)
        # Null out the signals that are less than the mean,
        patch[patch < patch.mean()] = 0
        # Starting from the center of the patch, slide to the right
        # as long as the signal isn't 0. Helps adjust the patch center
        # if one of the left lane ends up extending out the center of the patch.
        for i in range(int(len(patch)/2), len(patch)):
            n = i
            if patch[i] == 0: break
                
        # If get to the end of the image and no dip in the signal was seen then 
        # bail out - rest of the algorithm won't work
        if n == (len(patch)-1): return [],[], n

        # Things look good if we're here. Find the left / right peak points and report result
        fig, axs = plt.subplots(1, 2, sharey=True) if show_plots else (None, [None]*2)
        left, l_winx = find_peaks(patch[:n], centers[0], axs=axs[0])
        right,r_winx =find_peaks(patch[n:], centers[1]-n, axs=axs[1], offset=n)        
        if show_plots: plt.show()
        return left, right, n, l_winx, r_winx

    
    def find_lane_markers(be_sbl, num_seqs=16, show_plots=False):
        ''' Main routine that finds the lane markers. 
            Returns: BirdsEye image with lane markers overlayed
        '''
        
        # Acquire image shape
        h, w = be_sbl.shape

        step = int(h/num_seqs) # PARAM_TUNING window height of the vertical scan patch
        skip = 4               # PARAM_TUNING number of horizontal pixels to skip
        
        # Util functions
        norm = normalize
        downsample = lambda a, s=3: a[:s*(len(a)//s)].reshape((len(a)//s), s).sum(axis=1)

        # Kernel size of patch filter
        ksize = 31 # PARAM_TUNING
        filt = cv2.getDerivKernels(1, 0, ksize)
        # Differential filter and a mean filter
        div, box = norm(filt[0].reshape(ksize)), norm(gaussian(ksize, ksize/3))
        kernel = div * box
        
        im = None
        left_pts, right_pts, mid_pts = [], [], []
        # Using udacity's sample code to estimate the global center of the image.
        l_center, r_center = uda.find_centers(be_sbl)
        centers = []
              
        # Scan down the image vertically in step size of patch height
        for i in range(h, 0, -step):
            # Get the patch
            patch = be_sbl[(i-step):i, :]
            # Sum-up patch vertically
            patch_sum = patch.sum(axis=0)
            if skip > 1: patch_sum = downsample(patch_sum, skip)
                    
            # Calculate 1-D patch signal derivative 
            patch_dif = np.convolve(patch_sum, kernel, mode='same')
            # Calculate 1-D patch signal's guassian sampled sum
            patch_sum = np.convolve(patch_sum, box,    mode='same')
            
            # PARAM_TUNING: If no global/local centers are known than just use the hard-coded locations
            if len(centers) == 0: centers = np.array([(l_center, frame.shape[0]*(7.0/8.0)), (r_center, frame.shape[0]*(7.0/8.0))], dtype=np.int)

            # Find left and right lane peak points
            p_centers = (np.array(centers,)[:,0] / skip).astype(np.int)
            left, right, n, l_winx, r_winx = find_left_right_peaks(patch_sum, p_centers)
            l_winx = np.array(l_winx, dtype=np.int) * skip
            r_winx = np.array(r_winx, dtype=np.int) * skip
            
            # Append the results to left and right points accumulator array
            y_pt = int(np.mean([i, i-step]))
            if (len(left) > 0): 
                left   = [int(left[0]*skip), y_pt]
                left_pts.append(left)
                centers[0] = left
            if (len(right) > 0): 
                right  = [int(right[0]*skip), y_pt]
                right_pts.append(right)
                centers[1] = right

            # Draw the recovered lane points on the image (visualization purpose only)
            if im is None: im = getColor(be_sbl*255)
            im = rect(im, (0,i-step), (w, step), copy=False)
            im = rect(im, (l_winx[0],i-step), (l_winx[1] - l_winx[0], step), color=[0,255,255], copy=False)
            im = rect(im, (r_winx[0],i-step), (r_winx[1] - r_winx[0], step), color=[0,255,255], copy=False)
            im = circle(im,  left_pts, color=[255,0,0], copy=False)
            im = circle(im, right_pts, color=[255,0,0], copy=False)

            if show_plots: 
                plt.imshow(im)
                plt.show()

        ## DONE SCANNING THROUGH THE IMAGE (VERTICAL)
                
        # Fit a second order polynomial to each left and right lane points
        curve_l, curve_r = fit_lane_to_curves(left_pts), fit_lane_to_curves(right_pts)
        conf_l,  conf_r  = calc_fit_confidence(curve_l, left_pts), calc_fit_confidence(curve_r, right_pts)
        # Use the quadratic fits to generate lane points
        curve_pts_l, curve_pts_r = create_curve(curve_l, h), create_curve(curve_r, h)
        # Reference Lane Selection. This is the lane we have most confidence on.
        # Reference lanes are selected based on - the lane that has most recovered lane markers
        ref_curve = 'left' if (conf_l[0] > conf_r[0]) else 'right'
        # If no lanes have enough confidence then just set reference lanes to None
        ref_curve = None if (conf_l[0] < 0.5 and conf_r[0] < 0.5) else ref_curve
    
        # PARAM_TUNING
        lane_err_max = 2000.
        err_thresh_1 = 0.45
        err_thresh_2 = 0.70
        err_thresh_3 = 0.85
        err_thresh   = err_thresh_1
        
        def choose_curve(curves, ref_curve):
            ''' Given a bunch of quadratic curves (potential choices), 
                this routine returns the best choice curve from the list
                that minimizes the x**2 difference between the reference curve.
                
                Returns : best choice curve or None if no reference curve is provided
            '''
            if ref_curve is None: return 0
            # Array of x**2 coefficients 
            p2 = np.array([c[0] for c in curves])
            # Calculate the sum-of-squared difference between curves and refcurve 
            p2diff = np.absolute(p2 - ref_curve[0])
            # Return the curve that minimizes the difference
            ind = np.argsort(p2diff)
            return ind[0]

        def run_history(hist, curve, curve_pts, ref_curve, conf_err=None):
            ''' History Tracking '''                
            if (len(curve) < 3):
                # If non-quadratic curve was provided, then use the 
                # avg curve from history
                avg = hist.get_avg_coeffs()
                if (avg is None): return None, curve
                hist.add_coeffs(avg)              
                return None, avg
            else:
                err, update, errs = None, True, np.array([0., 0., 0.])
    
                # Useful curve provide - do rest of the logic.
                cur_avg_curve = None
                org_curve = curve                   # Curved provided as input arg
                avg_curve = hist.get_avg_coeffs()   # Average curve from history
                prvs = hist.get('cpts')             # Get the previous curve points. None if no history
                
                # Check if this the first time - means no history - otherwise do nothing
                if prvs is not None and avg_curve is not None:

                    cpts = curve_pts[::4]               # PARAM_TUNING: Curve points passed in, subsampled by 4 
                    avg_pts = create_curve(avg_curve, h)[::4] # Avg curve points

                    # Calculate the euclidean distance difference between  
                    # corresponding avg_points and current curve points (SMALLER IS BETTER)
                    lane_err = min(1.0, (np.absolute(np.array(cpts)[:,0] - np.array(avg_pts)[:,0]).sum())/lane_err_max)
                    err = np.sum([lane_err, conf_err])
                    
                    avg_curve = avg_curve 
                    cur_avg_curve = [(curve[i]+avg_curve[i])/2.0 for i in range(3)]
                    
                    # PARAM_TUNING: weighted_avg
                    wavg = normalize(np.array([1., .5, .25, .125, 0.]))
                    def weighted_avg(vec, w=wavg):
                        n = min(len(vec), len(w))
                        return np.sum(vec[:n] * w[:n])

                    prev_curves = np.array(hist.get_coeffs()[:3])
                    cur_avg_curve = [ weighted_avg(np.append(curve[i], prev_curves[:,i])) for i in range(3) ]
                    
                    if err > err_thresh: 
                        # Choices for curves:
                        # 1. Current Curve
                        # 2. Avg Curve
                        # 3. Mean(Current Curve, Avg Curve)
                        if err < err_thresh_2:  
                            # Level-1 error: correctable
                            curves = [cur_avg_curve, avg_curve, curve] 
                        elif err < err_thresh_3:
                            # Level-2 error: non-correctable
                            curves = [avg_curve, cur_avg_curve]                            
                        else:
                            # Level-3 error: pretty-bad 
                            curves = [avg_curve]

                        # Find the best curve that minimizes difference between the ref_curve
                        best_ind = choose_curve(curves, ref_curve)
                        curve = curves[best_ind] #if best_ind is not None else cur_avg_curve                                            
                    
                    errs = np.array([err, lane_err, conf_err])
#                     if np.all(curve == avg_curve): update = False
                        
                if update:
                    # Create new curve points based on above result, could be the orginial points
                    cpts = create_curve(curve, h)[::4]
                    hist.add('cpts', cpts) # Update curve-points history
                    hist.add_coeffs(curve) # Update curve-params history          

                return [err, avg_curve, cur_avg_curve, org_curve, errs], curve
        
        def plot_history(axs, hist, nsamples=50):
            avgs, errs = hist.get('c_avg'), hist.get('c_err')
            if avgs is None or errs is None: return
            n = min(min(len(avgs), len(errs)), nsamples)
            y = np.zeros(n)
            x = np.array(avgs[:n]).reshape(n,3)
            for i, c in enumerate(['r--', 'g--', 'b--']): axs[i].plot(x[:,i], c)                
            axs[3].plot(errs[:n])
             
        if ref_curve == 'right':
            r_err, curve_r = run_history(r_hist, curve_r, curve_pts_r, None,    conf_err=(1.0 - conf_r[0]))
            l_err, curve_l = run_history(l_hist, curve_l, curve_pts_l, curve_r, conf_err=(1.0 - conf_l[0]))
        else:
            ref = curve_l if ref_curve is not None else None
            l_err, curve_l = run_history(l_hist, curve_l, curve_pts_l, None, conf_err=(1.0 - conf_l[0]))
            r_err, curve_r = run_history(r_hist, curve_r, curve_pts_r, ref, conf_err=(1.0 - conf_r[0]))
        
        curve_pts_l, curve_pts_r = create_curve(curve_l, h), create_curve(curve_r, h)

        if draw_hist: 
            fig, axs = plt.subplots(4,2) #sharey=True, sharex=True, 
            if l_err is not None: plot_history(axs[:,0], l_hist)
            if r_err is not None: plot_history(axs[:,1], r_hist)
            plt.show()    

    
        line_clr = [None, magenta, blue_l, yellow, white]
        line_nms = [None, "Avg", "Cvg", "Org", "Ers"]
        # Draw all the debug lines and text
        img = np.zeros((h, w, 3), dtype=np.uint8)
        for err in [l_err, r_err]:
            if err is None: continue
            for i in [1,2,3]:
                if (err[i] is not None): img = circle(img, create_curve(err[i], h), radius=12, color=line_clr[i],copy=False)
                    
        img = circle(img, curve_pts_l, radius=12, color= green,  copy=False)
        img = circle(img, curve_pts_r, radius=12, color= green,  copy=False)
        img = circle(img, left_pts,    radius=10, color= red,    copy=False)
        img = circle(img, right_pts,   radius=10, color= red,    copy=False)  
        img = circle(img, centers,     radius=10, color= blue_l, copy=False)

        def printError(img, err, conf, crv, loc, lclr=line_clr, lnms=line_nms):
            txt = []
            if (err is not None and err[0] is not None):
                if (err[0] > err_thresh): clr = red
                txt+= [["Err: {:1.3f} Conf: {:1.2f} {}".format(err[0], conf, ref_curve), green_l if (err[0] < err_thresh) else red]]
                txt+= [["Cur: [ {:1.4f}, {:1.4f}, {:3.1f}]".format(*list(crv)), green_l]] ## Current Curve                
                for i in range(1, len(err)):
                    if (err[i] is not None): txt+= [["{:3s}: [ {:1.3f}, {:1.3f}, {:1.3f} ]".format(lnms[i], *list(err[i])), lclr[i]]]

            for i, t in enumerate(txt):
                img = cv2.putText(img, t[0], (loc[0], loc[1]+30*i), cv2.FONT_HERSHEY_PLAIN, 2, color=t[1], thickness=3)
            return img
        
        l_err += conf_l[1:]
        r_err += conf_r[1:]
        line_nms += ["Cnf"]
        line_clr += [white]
                
        txt = np.zeros_like(ud_frame)
        txt = cv2.putText(txt, "Frame# " + str(frame_num), (50, 50), cv2.FONT_HERSHEY_PLAIN, 3, color=white, thickness=3)
        txt = printError(txt, l_err, conf_l[0], curve_l, ( 50,100))
        txt = printError(txt, r_err, conf_r[0], curve_r, (700,100))

        points, curves = (left_pts+right_pts), (curve_pts_l+[p for p in reversed(curve_pts_r)])
        return img, im, txt, points, curves, [curve_l, curve_r]

    
    def calc_curv_rad_and_center_dist(bin_img, l_fit, r_fit, l_lane_inds, r_lane_inds):
        # Define conversions in x and y from pixels space to meters
        ym_per_pix = 3.048/100 # meters per pixel in y dimension, lane line is 10 ft = 3.048 meters
        xm_per_pix = 3.7/378 # meters per pixel in x dimension, lane width is 12 ft = 3.7 meters
        left_curverad, right_curverad, center_dist = (0, 0, 0)
        # Define y-value where we want radius of curvature
        # I'll choose the maximum y-value, corresponding to the bottom of the image
        h = bin_img.shape[0]
        ploty = np.linspace(0, h-1, h)
        y_eval = np.max(ploty)

        # Identify the x and y positions of all nonzero pixels in the image
        nonzero = bin_img.nonzero()
        nonzeroy = np.array(nonzero[0])
        nonzerox = np.array(nonzero[1])
        # Again, extract left and right line pixel positions
        leftx = nonzerox[l_lane_inds]
        lefty = nonzeroy[l_lane_inds] 
        rightx = nonzerox[r_lane_inds]
        righty = nonzeroy[r_lane_inds]

        if len(leftx) != 0 and len(rightx) != 0:
            # Fit new polynomials to x,y in world space
            left_fit_cr = np.polyfit(lefty*ym_per_pix, leftx*xm_per_pix, 2)
            right_fit_cr = np.polyfit(righty*ym_per_pix, rightx*xm_per_pix, 2)
            # Calculate the new radii of curvature
            left_curverad = ((1 + (2*left_fit_cr[0]*y_eval*ym_per_pix + left_fit_cr[1])**2)**1.5) / np.absolute(2*left_fit_cr[0])
            right_curverad = ((1 + (2*right_fit_cr[0]*y_eval*ym_per_pix + right_fit_cr[1])**2)**1.5) / np.absolute(2*right_fit_cr[0])
            # Now our radius of curvature is in meters

        # Distance from center is image x midpoint - mean of l_fit and r_fit intercepts 
        if r_fit is not None and l_fit is not None:
            car_position = bin_img.shape[1]/2
            l_fit_x_int = l_fit[0]*h**2 + l_fit[1]*h + l_fit[2]
            r_fit_x_int = r_fit[0]*h**2 + r_fit[1]*h + r_fit[2]
            lane_center_position = (r_fit_x_int + l_fit_x_int) /2
            center_dist = (car_position - lane_center_position) * xm_per_pix
        return left_curverad, right_curverad, center_dist
    print('...')    

    
    blend = lambda imgA, imgB, a=0.5, b=0.5, dtype=np.uint8: (a*imgA + b*imgB).astype(dtype)
    
    ## Undistort
    ud_frame = undistort(frame, calib_matrix, calib_dist, retRoi=False)    

    ## Apply Bird's Eye Transformation
    be_img, mat, inv_mat, src, dst = BirdsEye.TransformImage(ud_frame, pts=be_pts, retMats=True, retPts=True)
    
    ## After Sobel-Magnitude-Thresholding
    be_sbl = Sobel.Mag_thresh(be_img, *sbl_param)
    
    ## Make a mask vthat covers the destination quad
    be_mask_pts = widen(dst, left=50, right=150)
    be_mask_pts = shorten(be_mask_pts, top=-50, bot=-10)
    be_mask = create_region_mask(be_sbl.shape, np.array([be_mask_pts], dtype=np.int32)).astype(bool)
    be_sbl[be_mask == False] = 0
    
    ## Show images until now
    if show_plots: renderAllImages(frame, ud_frame, be_img, be_sbl, src, dst, be_mask_pts)    
        
    ## Find lane lines
    res = find_lane_markers(be_sbl, show_plots=show_plots)
    img, tmp_img, figimg, points, curves, coeffs = res
    
    ## Project the output image on to the road
    img_masked    = apply_mask(img, be_mask, false_value=[170, 0, 100])
    img_projected = transformImage(img_masked, inv_mat)

    ## Project Points on undistorted frame
    points = transformPoints(points, inv_mat).astype(np.int)
    curves = transformPoints(curves, inv_mat).astype(np.int)
    ud_img = np.zeros_like(ud_frame)
    ud_img = polyfills(ud_img, curves, color=[0,  255, 80, 0], copy=False)
    ud_img = circle(ud_img, points, radius=4, color=[255, 0,   0], copy=False)
    ud_img = blend( ud_img, figimg, a=1, b=1)
    ud_img = blend(ud_frame, ud_img, a=0.7, b=0.3)

    be_sbl_lines = blend(img, be_img)
#     be_sbl_lines = blend(img, getColor(be_sbl*255))
    be_sbl_lines = blend(figimg, be_sbl_lines, a=0.4, b=1.)

    im1, im2, im3, im4 = be_sbl_lines, tmp_img, ud_img, figimg
#     print(" {}, {}, {}, {} ".format(im1.shape, im2.shape, im3.shape, im4.shape))

    return np.vstack((np.hstack((imresize(im1, 0.5), imresize(im2, 0.5))), im3))
#     return np.vstack((np.hstack((im1,im2)), np.hstack((im3,im4))))
#     return np.hstack((im3, im4))
#     return np.vstack((im1, im3))
#################################### RUN ####################################

_Pts = road_trapezoid(3)
# _Pts = [(284, 654), (624, 436), (682, 437), (1041, 651)]
# _Pts = [(298, 649), (614, 441), (687, 441), (1033, 639)]
# _Pts = [(284, 654), (616, 441), (690, 442), (1041, 651)]

sbl_param=(15, (10, 255))
l_hist, r_hist = History(), History()

# run_video(start=0, fheight=10, func=doAll, be_pts=_Pts, sbl_param=sbl_param, l_hist=l_hist, r_hist=r_hist, cmap='gray')

# test_frame = vid.get_data(384)
# img = doAll(test_frame, be_pts=_Pts, sbl_param=sbl_param, l_hist=l_hist, r_hist=r_hist, verbose=True, show_plots=True, draw_hist=False);
# plt.imshow(img), plt.show;


l_hist, r_hist = History(4), History(4)
for i in np.arange(965, 1100):
    draw_hist = (i in range(550, 1250))    
    test_frame = vid.get_data(i)
    img = doAll(test_frame, frame_num=i, be_pts=_Pts, sbl_param=sbl_param, l_hist=l_hist, r_hist=r_hist, verbose=False, show_plots=False, draw_hist=False);
    
    if draw_hist:
        fig, axs = subplots(1,1, figsize=(10,10))
        axs.imshow(img);
        axs.set_title("Frame: "+str(i))
        plt.show();
    else:
        print("i={} draw_hist = {}".format(i, draw_hist))
    clear_output(wait=True)
```
